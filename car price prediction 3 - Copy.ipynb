```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe=OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_,handle_unknown="ignore"),['name','company','fuel_type']),remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred= pipe.predict(x_test)
```

```python
y_pred
```

```python
r2_score(y_test,y_pred)
```

```python
scores=[]
for i in range(1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(x_train,y_train)
    y_pred=pipe.predict(x_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]# get max value of r2
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=433)
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(x_train,y_train)
y_pred1=pipe.predict(x_test)
r2_score(y_test,y_pred1)
```

```python
 from sklearn.ensemble import BaggingRegressor
```

```python
# now we apply bagging on it using  pipeline and diff random state of x train then se what we get max r2 then we apply that one 
Scores=[]
for i in range(1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
   
    br=BaggingRegressor(estimator=LinearRegression(),n_estimators=50, random_state=42)
    pipe_br=make_pipeline(column_trans,br)

    pipe_br.fit(x_train, y_train)
    y_pred2 = pipe_br.predict(x_test)

    Scores.append(r2_score(y_test, y_pred2))
    
    



```

```python
np.argmax(Scores)
```

```python
# we get best result at random sstate 124 so now we do all model on 124 so we easily compare 
```

```python
# if we need scaling then we use it object in code 
from sklearn.preprocessing import StandardScaler
ss=StandardScaler(with_mean=False)
```

```python
from sklearn.linear_model import Ridge
r=Ridge(alpha=1)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=124)
pipe_r=make_pipeline(column_trans,ss,r)
pipe_r.fit(X_train, Y_train)
Y_pred5 = pipe_r.predict(X_test)
r2_score(Y_test, Y_pred5)
```

```python
from sklearn.linear_model import Lasso
l=Lasso(alpha=150, max_iter=5000)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=124)
pipe_l=make_pipeline(column_trans,ss,l)
pipe_l.fit(X_train, Y_train)
Y_pred6 = pipe_l.predict(X_test)
r2_score(Y_test, Y_pred6)
```

```python
 from sklearn.ensemble import RandomForestRegressor
```

```python
# now we apply random forest on it 
 

X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=124)
rfr=RandomForestRegressor( n_estimators=80,random_state=42)
pipe_rfr=make_pipeline(column_trans,rfr)
pipe_rfr.fit(X_train, Y_train)
Y_pred3 = pipe_rfr.predict(X_test)
r2_score(Y_test, Y_pred3)
# so in random forest regressionn wwith random state 124 we achive good result

   


```

```python
# now we perforem knn also 
from sklearn.neighbors import KNeighborsRegressor
knnr=KNeighborsRegressor(n_neighbors=20)
pipe_knnr=make_pipeline(column_trans,ss,knnr)
pipe_knnr.fit(X_train, Y_train)
Y_pred4 = pipe_knnr.predict(X_test)
r2_score(Y_test, Y_pred4)
```
